# League Visualization Project

```python
# importing all the packages
import re
import numpy as np
import pandas as pd
from selenium import webdriver
import time 
from bs4 import BeautifulSoup
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
browser = webdriver.Firefox(executable_path='/Users/Allen/desktop/UVA/HackCville/Node A/geckodriver')
browser.get('https://matchhistory.na.leagueoflegends.com/en/#match-history/NA1/226687343?matchType=2&map=1')
```

```python
table = browser.find_elements_by_class_name("kda-plate-kda")

for i in table:
    entry = i.get_attribute('outerHTML')
    print(entry)
```

```python
table = browser.find_elements_by_class_name("kda-plate-kda") # used to find all the kda values of each champion 
kda_finder = re.compile('>([0-9]+)<')
kda_list = []

for i in table:
    entry = i.get_attribute('outerHTML')
    
    for match in kda_finder.finditer(entry):
        kda_list.append(match.group(1))

print(kda_list)

kills = []
deaths = []
assists = []
for i in range(0, len(kda_list), 3):
    kills.append(kda_list[i])
print(kills)
for i in range(1, len(kda_list), 3):
    deaths.append(kda_list[i])
print(deaths)
for i in range(2, len(kda_list), 3):
    assists.append(kda_list[i])
print(assists)
```

```python
table = browser.find_elements_by_class_name("champion-nameplate-name") # used to find all the champion names 
champion_name = re.compile('style="">([A-Z][a-z]+)<') 
champion_list = []

for i in table:
    entry = i.get_attribute('outerHTML')

    for match in champion_name.finditer(entry):
        champion_list.append(match.group(1))

print(champion_list)
```

```python
table = browser.find_elements_by_class_name("date-duration-duration") # used to find total game time 
game_time = re.compile('>([0-9]+[:][0-9]+)<')
time = []

for i in table:
    entry = i.get_attribute('outerHTML')

    for match in game_time.finditer(entry):
        match = match.group(1)
        match = match.replace(':', '.')
        if len(match) == 5 and int(match[3]) != 0:
            match = str(int(match[0:2]) + int(match[-2:])/60)
        elif int(match[3]) == 0: 
            match = str(float(match[0:4]) + int(match[-1:])/60)
        else:
            match = str(int(match[0:1]) + int(match[-2:])/60)
        time.append(round(float(match), 2))
        
print(time)
```

```python
table = browser.find_elements_by_class_name("content-border") # used to find whether I won or lost the game
game_result = re.compile('game-summary-([a-z]+)')
result = []

for i in table:
    entry = i.get_attribute('outerHTML')
    
    for match in game_result.finditer(entry):
        result.append(match.group(1))

print(result)
```

```python
game_result = pd.Series(result)
game_time = pd.Series(time)
champion_name = pd.Series(champion_list)

league_stats = pd.DataFrame({"Result": game_result, 'Game time':game_time, 'Champion': champion_name})

league_stats['Kills'] = pd.Series(kills)
league_stats['Deaths'] = pd.Series(deaths)
league_stats['Assists'] = pd.Series(assists)

league_stats
```

```python
league_stats.info()
league_stats.describe()
```

```python
objects = ('Victory', 'Defeat')
y_pos = np.arange(len(objects))
result = [(28/55) * 100, (27/55) * 100]

plt.bar(y_pos, result, align = 'center', alpha=0.2)
plt.xticks(y_pos, objects)
plt.ylabel('Percentage')
plt.title('Win-loss percentage on Braum')

plt.show()
```

```python
league_stats['Kills'] = pd.to_numeric(league_stats['Kills'])
league_stats['Assists'] = pd.to_numeric(league_stats['Assists'])

sns.lmplot(data = league_stats, x = 'Kills', y = 'Assists', hue = 'Champion', fit_reg = True, legend = True)
plt.show()
```
